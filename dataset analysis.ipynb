```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_obj(name):
    with open(name, 'rb') as f:
        return pickle.load(f)
```

```python
ml1 = load_obj('ml-1m.pkl')
ml10 = load_obj('ml-10m.pkl')
dob = load_obj('utils/douban.pkl')
flix = load_obj('utils/flixster.pkl')

for df in (ml1,ml10,dob,flix):
    df.columns = ['user_id','item_id','rating','partition']
```

```python
dfs = (ml1,ml10,dob,flix)
```

```python
dfs_df = pd.DataFrame()
dfs_df['attribute'] = ['ml-1m','ml-10m','douban','flixster']
dfs_df['number_of_ranks'] = [len(df) for df in dfs]
dfs_df['number_of_users'] = [df['user_id'].nunique() for df in dfs]
dfs_df['number_of_items'] = [df['item_id'].nunique() for df in dfs]
dfs_df['fill_ratio'] = dfs_df['number_of_ranks']/(dfs_df['number_of_users']*dfs_df['number_of_items'])
dfs_df['fill_ratio'] = dfs_df['fill_ratio'].round(4)
dfs_df.head()
```

```python
dfs_df['mean rnaks per user'] = [df.groupby('user_id').size().mean().round(4) for df in dfs]
dfs_df['median rnaks per user'] = [df.groupby('user_id').size().median() for df in dfs]
dfs_df['max rnaks_per user'] = [df.groupby('user_id').size().max() for df in dfs]
dfs_df['min rnaks_per user'] = [df.groupby('user_id').size().min() for df in dfs]

dfs_df['mean rnaks per item'] = [df.groupby('item_id').size().mean().round(4) for df in dfs]
dfs_df['median rnaks per item'] = [df.groupby('item_id').size().median() for df in dfs]
dfs_df['max rnaks per item'] = [df.groupby('item_id').size().max() for df in dfs]
dfs_df['min rnaks per item'] = [df.groupby('item_id').size().min() for df in dfs]


for i in range(1,21):
    dfs_df[f'items_apears_{i}_times'] = [(df.groupby('item_id').size()==i).sum() for df in dfs]
```

```python
data_summery_df = dfs_df.head().T.reset_index()
data_summery_df.columns = data_summery_df.iloc[0]
data_summery_df.drop(data_summery_df.index[0],inplace=True)
data_summery_df
```

```python
data_summery_df.to_csv('data_summery_df.csv',index=False)
```
